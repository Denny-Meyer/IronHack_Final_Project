```python
import matplotlib.pyplot as plt
```

```python
data = []
```

```python
with open('rewards.txt','r') as f:
    lines = f.readlines()
    for num in lines:
        if float(num) < -10:
            data.append(-1)
        else:
            data.append(float(num))
```

```python
# number of steps
len(data)
```

```python
plt.figure(figsize=(40,20))
plt.plot(data)
```
